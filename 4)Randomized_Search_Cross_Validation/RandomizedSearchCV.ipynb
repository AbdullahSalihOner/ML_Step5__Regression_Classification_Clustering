```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.model_selection import train_test_split,RandomizedSearchCV  
from sklearn.preprocessing import StandardScaler    
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('kanser.csv')    
data.head()
```

```python
data.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis] # M=1, B=0
```

```python
data.head()
```

```python
y = data.diagnosis.values # diagnosis sütununu y değişkenine atadık // we assign the diagnosis column to the y variable
x = data.drop(['diagnosis'],axis=1) # diagnosis sütununu x değişkenine atadık // we assign the diagnosis column to the x variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.20,random_state=33) # veri setini train ve test olarak ayırdık // we split the dataset into train and test
```

```python
sc = StandardScaler() # verileri ölçeklendirmek için kullanıyoruz   // we use it to scale the data  
xtrain = sc.fit_transform(x_train) # train verilerini ölçeklendirdik  // we scale the train data  
xtest = sc.transform(x_test) # test verilerini ölçeklendirdik   // we scale the test data 
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=40) # modeli tanımladık // we defined the model
```

```python
param_random = {
    "n_estimators":[25,50,75,100,150,200],  
    "max_depth":[5,8,12,15,20,22],
}
```

```python
rm = RandomizedSearchCV(model,param_random,cv=5,n_iter=10,scoring="accuracy",n_jobs=-1,random_state=23) # modeli eğittik // we trained the model
rm.fit(xtrain,y_train) 
```

```python
print("En iyi parametreler : "+str(rm.best_params_)) 
print("En iyi skor : "+str(rm.best_score_))
```

```python
best_test = rm.best_estimator_ # en iyi parametrelerle modeli oluşturduk // we created the model with the best parameters
accuracy = best_test.score(xtest,y_test)    
```

```python
print("Test verileri için doğruluk oranı : "+str(accuracy)) # we found the accuracy rate for the test data / doğruluk oranını bulduk
```
