```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('musteri.csv')  
```

```python
data.head()
```

```python
x = data.iloc[:, [3, 4]].values
```

```python
import scipy.cluster.hierarchy as sch
den= sch.dendrogram(sch.linkage(x, method = 'ward'))    # ward methodu ile dendogram oluşturduk // we created dendogram with ward method
```

```python
from sklearn.cluster import AgglomerativeClustering # agglomerative clustering kütüphanesini import ettik // we imported agglomerative clustering library
ac = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage ='ward')  # ward methodu ile 5 cluster oluşturduk // we created 5 cluster with ward method
yhead = ac.fit_predict(x)  # x verilerini fit ettik  // we fit x data
```

```python
plt.scatter(x[yhead == 0, 0], x[yhead == 0, 1], s = 100, c = 'red', label = 'Cluster 1')  # cluster 1
plt.scatter(x[yhead == 1, 0], x[yhead == 1, 1], s = 100, c = 'blue', label = 'Cluster 2') # cluster 2
plt.scatter(x[yhead == 2, 0], x[yhead == 2, 1], s = 100, c = 'green', label = 'Cluster 3') # cluster 3
plt.scatter(x[yhead == 3, 0], x[yhead == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')  # cluster 4
plt.scatter(x[yhead == 4, 0], x[yhead == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') # cluster 5

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
